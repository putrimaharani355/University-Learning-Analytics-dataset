```python
encoder = LabelEncoder()
result['assessment_type'] = encoder.fit_transform(result['assessment_type'])
result['region'] = encoder.fit_transform(result['region'])
result['highest_education'] = encoder.fit_transform(result['highest_education'])
result['imd_band'] = encoder.fit_transform(result['imd_band'])
result['age_band'] = encoder.fit_transform(result['age_band'])
result['disability'] = encoder.fit_transform(result['disability'])
result['final_result'] = encoder.fit_transform(result['final_result'])
result['gender'] = encoder.fit_transform(result['gender'])
result.head()
```

# Pembagian antara features dan labels


```python
features = result.drop(columns=['final_result'])
labels = result['final_result']
```

# Membagi train set sebesar 80% dan test set sebesar 20%

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.20, random_state=70)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

# Membuat model Decision Tree
clf = DecisionTreeClassifier()

# Melatih model Decision Tree pada data latih
clf.fit(X_train, y_train)
```

# Model CNN

```python
scaler = MinMaxScaler()
features= scaler.fit_transform(features)
```

```python
features= features.reshape(-1, features.shape[1], 1, 1)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Conv1D(filters=37, kernel_size=3, activation='relu', input_shape=(X_train.shape[1], 1)),
    tf.keras.layers.MaxPooling1D(pool_size=2),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Conv1D(filters=64, kernel_size=3, activation='relu'),  # Tambahkan lapisan Conv1D ini
    tf.keras.layers.MaxPooling1D(pool_size=2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(4, activation='softmax')
])
```

```python
# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model 2
model.fit(X_train, y_train, epochs=25, batch_size=64)
```

```python
# Evaluate the model on test data
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f'Test loss: {loss:.4f}, Test accuracy: {accuracy:.4f}')
```

```python
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=-1)
ytrue = y_test.astype(int).tolist()
```

```python
target_names = ['Class 0', 'Class 1', 'Class 2', 'Class 3']
print(classification_report(y_test, y_pred_classes, target_names=target_names))
```
